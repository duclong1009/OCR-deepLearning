```python
from glob import glob
import os
import json
import numpy as np

data_dir = "/content/drive/MyDrive/BK_AI/Dataset/VinText"
gt_dir = "/content/drive/MyDrive/BK_AI/Dataset/VinText/gt_test"
img_dir = "/content/drive/MyDrive/BK_AI/Dataset/VinText/unseen_test_images"


def load_gt(gt_path):
    lines = open(gt_path, encoding='utf-8').readlines()
    bboxes = []
    words = []
    res = []
    for line in lines:
        ori_box = line.strip().encode('utf-8').decode('utf-8-sig').split(',')
        box = [int(ori_box[j]) for j in range(8)]
        word = ori_box[8:]
        word = ','.join(word)
        # print(word)
        box = np.array(box, np.int32).reshape(4, 2)
        bboxes.append(box)
        words.append(word)

    for i in range(len(bboxes)):
        res.append({
            "transcription": words[i],
            "points": bboxes[i].tolist()
        })

    return res


def make_label_for_paddleocr():
    gt_paths = []
    img_paths = []

    img_names = os.listdir(img_dir)
    gt_names = os.listdir(gt_dir)
    gt_names = sorted(gt_names)

    for gt_name in gt_names:
        img_name = gt_name.replace("gt_", "").replace(".txt", ".jpg")
        img_path = "unseen_test_images/" + img_name

        if img_name not in img_names:
            print("Have", img_name)
            continue

        img_paths.append(img_path)
        gt_paths.append(os.path.join(gt_dir, gt_name))
    
    for idx, gt_path in enumerate(gt_paths[:200]):
        img_path = img_paths[idx]
        res = load_gt(gt_path)
        print(res)

        with open(os.path.join(data_dir, "label.txt"), "a", encoding="utf-8") as f:
            f.write(img_path + "\t")

            json.dump(res, f)

            f.write("\n")
        
make_label_for_paddleocr()
```
