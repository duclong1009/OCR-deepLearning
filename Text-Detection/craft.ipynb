```python
import json
iou = []
precision = []
recall = []
hmean = []

with open("log.json", "r") as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip("\n")
        log = json.loads(line)
        iou.append(log['macro_iou'])

with open("/content/drive/MyDrive/Project_Deeplearning/log2.json", "r") as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip("\n")
        if line == '':
            print(1)
            continue
        log = json.loads(line)
        precision.append(log['precision'])   
        recall.append(log['recall'])
        hmean.append(log['hmean'])

precision[0] = 0.79
recall[0] = 0.65
hmean[0] = 2.0*precision[0]*recall[0]/(precision[0] + recall[0])

iou[0], iou[1] = iou[1], iou[0]

plt.figure(figsize=(20, 7.5))
plt.subplot(1, 2, 1)
plt.plot([i for i in range(1, len(iou)+1)], iou, label="IoU")
plt.xlabel("Epoch")
plt.ylabel("IoU")

plt.subplot(1, 2, 2)
plt.plot([i for i in range(1, len(iou)+1)], precision, label="precision")
plt.plot([i for i in range(1, len(iou)+1)], recall, label="recall")
plt.plot([i for i in range(1, len(iou)+1)], hmean, label="hmean")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Metric")

plt.show()
print(iou[19])
print(precision[19])
print(recall[19])
print(hmean[19])
```

```python
log_txt = "/content/drive/MyDrive/Project_Deeplearning/2022_01_12_01_20_lr_0.0001/log.txt"
log_txt_2 = "/content/drive/MyDrive/Project_Deeplearning/2022_01_12_06_13_lr_0.00001"

log = []
loss = []
with open(log_txt, "r") as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip("\n")
        if line.startswith("epoch"):
            log.append(line)

num_per_epoch = [0 for _ in range(21)]
loss_per_epoch = [[0]*384 for _ in range(21)]
loss_last = []
for l in log:
    epoch = int(l.split("||")[0].split(":")[0].replace("epoch ", ""))
    if epoch == 0 or epoch == 21:
        continue
    loss = float(l.split("||")[2].strip(" ").replace("loss ", ""))
    iter = int(l.split("||")[0].split(":")[1].replace("(", "").replace("/384) ", ""))
    loss_per_epoch[epoch][iter] = loss
    loss_last.append(loss)
    num_per_epoch[epoch] += 1
print(num_per_epoch)
print(loss_last)
```

```python
import numpy as np
import matplotlib.pyplot as plt

tmp = []
indices = []
i = 0


for epoch in range(1, 21):
    # print("Num iter: ", num_per_epoch[epoch])
    loss_0_383 = loss_per_epoch[epoch]
    for i in range(0, 360, 40):
        loss_mean_20 = np.mean(np.array(loss_per_epoch[epoch][i:i+40], dtype=np.float32))
        tmp.append(loss_mean_20)
        indices.append(i+40+(epoch-1)*400)
        # print(i+20+(epoch-1)*400)
    
    last = np.mean(np.array(loss_per_epoch[epoch][360:], dtype=np.float32))
    
    tmp.append(last)
    indices.append(400*epoch)
print(indices)
print(tmp)
for i in range(len(indices)):
plt.plot(indices, tmp)
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()
```
